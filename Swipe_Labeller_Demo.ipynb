## Lets get you a dataset to label

In this case, a simple Airplane vs Not an Airplane dataset

```python
!gdown http://weegee.vision.ucmerced.edu/datasets/UCMerced_LandUse.zip
!rm -rf /content/UCMerced_LandUse
!unzip /content/UCMerced_LandUse.zip
```

```python
import os
import shutil
import random
from imutils import paths
```

```python
from imutils import paths
from PIL import Image
for img in list(paths.list_images('/content/UCMerced_LandUse/Images')):
  im = Image.open(img).convert('RGB').save(img.split('.')[0] + '.jpg', "JPEG", quality = 100)
  os.remove(img)
```

```python
all_images = []
all_images.extend(list(paths.list_images('/content/UCMerced_LandUse/Images/airplane')))
all_images.extend(list(paths.list_images('/content/UCMerced_LandUse/Images/beach')))
random.shuffle(all_images)
unlabelled_dataset = '/content/Unlabelled_Dataset'
if os.path.isdir(unlabelled_dataset):
  shutil.rmtree(unlabelled_dataset)
os.mkdir(unlabelled_dataset)
for i in range(len(all_images[:50])):
  shutil.copy(all_images[i],os.path.join(unlabelled_dataset,'{}.jpg'.format(i)))
shutil.rmtree('/content/UCMerced_LandUse')
```
